```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path = '/content/p_63060plot.csv'
data_p = pd.read_csv(file_path)

# Step 1: Calculate the mean of each row, ignoring the first column
row_means_p = data_p.iloc[:, 1:].mean(axis=1)

# Step 2: Plot the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=row_means_p, color='skyblue', orient='h')
plt.title('Box Plot of Row Means (Gene Expression)')
plt.xlabel('Mean Value')
plt.show()

# Calculate descriptive statistics for the row means
boxplot_stats = {
    "Mean": row_means_p.mean(),
    "Q1": row_means_p.quantile(0.25),
    "Median (Q2)": row_means_p.median(),
    "Q3": row_means_p.quantile(0.75),
    "Min": row_means_p.min(),
    "Max": row_means_p.max(),
}

# Display the box plot details
boxplot_stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path2 = '/content/h_63060plot.csv'
data_h = pd.read_csv(file_path2)

# Step 1: Calculate the mean of each row, ignoring the first column
row_means_h = data_h.iloc[:, 1:].mean(axis=1)

# Step 2: Plot the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=row_means_h, color='skyblue', orient='h')
plt.title('Box Plot of Row Means (Gene Expression)')
plt.xlabel('Mean Value')
plt.show()

# Calculate descriptive statistics for the row means
boxplot_stats = {
    "Mean": row_means_h.mean(),
    "Q1": row_means_h.quantile(0.25),
    "Median (Q2)": row_means_h.median(),
    "Q3": row_means_h.quantile(0.75),
    "Min": row_means_h.min(),
    "Max": row_means_h.max(),
}

# Display the box plot details
boxplot_stats

```

```python
# Transpose the data to prepare for box plot (samples as categories)
data_transposed = data_h.iloc[:, 1:].transpose()

# Plot the box plot for all samples
plt.figure(figsize=(15, 8))
sns.boxplot(data=data_transposed, color='skyblue')
plt.title('Box Plot of healthy')
plt.xlabel('Samples')
plt.ylabel('Gene Expression Values')
plt.xticks(rotation=90)
plt.show()
```

```python
import math

# Number of samples
num_samples = data_transposed.shape[1]

# Define the number of splits (4 plots)
num_splits = 4
samples_per_plot = math.ceil(num_samples / num_splits)

# Create multiple box plots
plt.figure(figsize=(15, 8 * num_splits))
for i in range(num_splits):
    plt.subplot(num_splits, 1, i + 1)
    start_idx = i * samples_per_plot
    end_idx = min((i + 1) * samples_per_plot, num_samples)
    sns.boxplot(data=data_transposed.iloc[:, start_idx:end_idx], color='skyblue')
    plt.title(f'Box Plot of healthy  {start_idx + 1} to {end_idx}')
    plt.xlabel('Samples')
    plt.ylabel('Gene Expression Values')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
