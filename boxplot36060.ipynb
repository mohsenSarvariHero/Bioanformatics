```python
import pandas as pd

# Reload the files
healthy_file_path = 'h_63060plot.csv'
patient_file_path = 'p_63060plot.csv'

# Reading the files into pandas DataFrames again
healthy_data = pd.read_csv(healthy_file_path)
patient_data = pd.read_csv(patient_file_path)

# Reshape the data for plotting
healthy_melted = healthy_data.melt(id_vars=["Unnamed: 0"], var_name="Sample", value_name="Expression")
healthy_melted["Group"] = "Healthy"

patient_melted = patient_data.melt(id_vars=["Unnamed: 0"], var_name="Sample", value_name="Expression")
patient_melted["Group"] = "Patient"

# Combine the datasets
combined_data = pd.concat([healthy_melted, patient_melted])

# Rename the gene identifier column for clarity
combined_data.rename(columns={"Unnamed: 0": "Gene"}, inplace=True)

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))
sns.boxplot(data=combined_data, x="Gene", y="Expression", hue="Group")
plt.title("Gene Expression Differences Between Healthy and Patient Samples (GSE36060)")
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Expression Level")
plt.xlabel("Gene")
plt.legend(title="Group")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File paths
healthy_file_path = 'h_63060plot.csv'
patient_file_path = 'p_63060plot.csv'
deg_file_path = 'h_deg_63060plot.csv'

# Reading the files into pandas DataFrames
healthy_data = pd.read_csv(healthy_file_path)
patient_data = pd.read_csv(patient_file_path)
deg_data = pd.read_csv(deg_file_path, header=None, names=["Gene"])

# Extract the DEGs into a set
deg_genes = set(deg_data["Gene"])

# Reshape the healthy and patient data for plotting
healthy_melted = healthy_data.melt(id_vars=["Unnamed: 0"], var_name="Sample", value_name="Expression")
healthy_melted["Group"] = "Healthy"

patient_melted = patient_data.melt(id_vars=["Unnamed: 0"], var_name="Sample", value_name="Expression")
patient_melted["Group"] = "Patient"

# Combine the datasets
combined_data = pd.concat([healthy_melted, patient_melted])
combined_data.rename(columns={"Unnamed: 0": "Gene"}, inplace=True)

# Mark genes with an asterisk if they are in the DEGs
combined_data["Gene_Annotated"] = combined_data["Gene"].apply(
    lambda x: f"{x}*" if x in deg_genes else x
)

# Create the boxplot
plt.figure(figsize=(14, 25))  # Larger height for better spacing
sns.boxplot(data=combined_data, y="Gene_Annotated", x="Expression", hue="Group", orient="h")
plt.title("Gene Expression Differences Between Healthy and Patient Samples (GSE36060)", fontsize=16)
plt.ylabel("Gene", fontsize=14)
plt.xlabel("Expression Level", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)

# Adjust subplot spacing to increase space between Y-axis elements
plt.subplots_adjust(left=0.2, right=0.95, top=0.95, bottom=0.05)
plt.legend(title="Group", fontsize=12, title_fontsize=14, loc="upper right")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File paths
healthy_file_path = 'h_63060plot.csv'
patient_file_path = 'p_63060plot.csv'
h_deg_file_path = 'h_deg_63060plot.csv'
p_deg_file_path = 'p_deg_63060plot.csv'

# Reading the files into pandas DataFrames
healthy_data = pd.read_csv(healthy_file_path)
patient_data = pd.read_csv(patient_file_path)
h_deg_data = pd.read_csv(h_deg_file_path, header=None, names=["Gene"])
p_deg_data = pd.read_csv(p_deg_file_path, header=None, names=["Gene"])

# Extract DEGs into sets
h_deg_genes = set(h_deg_data["Gene"])
p_deg_genes = set(p_deg_data["Gene"])
all_deg_genes = h_deg_genes.union(p_deg_genes)

# Reshape the healthy and patient data for plotting
healthy_melted = healthy_data.melt(id_vars=["Unnamed: 0"], var_name="Sample", value_name="Expression")
healthy_melted["Group"] = "Healthy"

patient_melted = patient_data.melt(id_vars=["Unnamed: 0"], var_name="Sample", value_name="Expression")
patient_melted["Group"] = "Patient"

# Combine the datasets
combined_data = pd.concat([healthy_melted, patient_melted])
combined_data.rename(columns={"Unnamed: 0": "Gene"}, inplace=True)

# Ensure all genes from h_deg and p_deg are included in the dataset
all_genes = pd.DataFrame({"Gene": list(all_deg_genes)})
all_genes["Expression"] = None  # Placeholder for missing values
all_genes["Group"] = "NA"  # To signify no group for these entries

# Merge the all_genes with the combined data
combined_data = pd.concat([combined_data, all_genes], ignore_index=True)

# Annotate genes with a '+' if they are in h_deg or p_deg
def annotate_gene(gene):
    if gene in all_deg_genes:
        return f"{gene}+"
    return gene

combined_data["Gene_Annotated"] = combined_data["Gene"].apply(annotate_gene)

# Create the boxplot
plt.figure(figsize=(14, 25))  # Larger height for better spacing
sns.boxplot(data=combined_data, y="Gene_Annotated", x="Expression", hue="Group", orient="h", dodge=True)
plt.title("Gene Expression Differences Between Healthy and Patient Samples (GSE36060)", fontsize=16)
plt.ylabel("Gene", fontsize=14)
plt.xlabel("Expression Level", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)

# Adjust subplot spacing to increase space between Y-axis elements
plt.subplots_adjust(left=0.3, right=0.95, top=0.95, bottom=0.05)
plt.legend(title="Group", fontsize=12, title_fontsize=14, loc="upper right")
plt.tight_layout()
plt.show()
```
